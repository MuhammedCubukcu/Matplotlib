## Import Libraries

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

## Scatter plot

Scatter plot, iki değişken arasındaki ilişkiyi görsel olarak gösteren istatistiksel bir grafiktir. Bu grafik,
veri setindeki her gözlem biriminin (örneğin, bir veri noktasının) belirli bir değere sahip olduğu iki farklı
değişken arasındaki ilişkiyi gösterir. Scatter plotun temel kullanımları şunlardır:

* İlişkiyi Gösterme: İki değişken arasındaki ilişkiyi gözlemlemek için kullanılır. Örneğin, x eksenindeki bir
değişkenin y eksenindeki bir değişkenle nasıl ilişkili olduğunu anlamak için scatter plot kullanabilirsiniz.

* Outlier Tespiti: Aykırı değerleri belirlemek için kullanışlıdır. Eğer veri setindeki bir gözlem diğerlerinden
belirgin bir şekilde farklı ise, bu scatter plot üzerinde dikkat çekecektir.

* Dağılımı Gösterme: Değişkenlerin dağılımını göstererek, veri setinin genel yapısal özelliklerini anlamanıza yardımcı olur.

* Korelasyon Analizi: İki değişken arasındaki ilişkinin doğasını belirlemek için kullanılabilir. Eğer değişkenler
arasında bir tür ilişki varsa, bu scatter plot üzerinde belirgin olacaktır.

* Regresyon Analizi: Eğer iki değişken arasındaki ilişkiyi modellemek ve gelecekteki değerleri tahmin etmek
istiyorsanız, scatter plot üzerinde regresyon doğrusu ekleyerek bu analizi yapabilirsiniz.

Özetlemek gerekirse, scatter plot, veri setindeki değişkenler arasındaki ilişkiyi görsel olarak anlamak ve analiz
etmek için kullanılan önemli bir araçtır.

```python
X_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
plt.scatter(X_data, y_data, color="#00008B", marker="*", s=100, alpha=0.9)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

## Line chart

Line chart, veri setindeki değerlerin zaman, sıralama veya başka bir sürekli değişken üzerindeki değişimini görsel
olarak gösteren bir grafiktir. Line chart'ın temel kullanımları şunlardır:

* Zamana Bağlı Değişimleri Gösterme: Line chart, belirli bir zaman aralığı içinde bir değişkenin nasıl değiştiğini
göstermek için kullanılır. Örneğin, bir hisse senedinin fiyatının bir yıl boyunca nasıl değiştiğini göstermek için
line chart kullanılabilir.

* Sıralamadaki Değişiklikleri Gösterme: Line chart, veri setindeki değerlerin sıralama veya sıradaki değişimini
göstermek için kullanılabilir. Örneğin, bir yarışmada katılımcıların sıralamasının zamanla nasıl değiştiğini
göstermek için kullanılabilir.

* İki veya Daha Fazla Değişkenin Karşılaştırılması: Birden fazla değişkenin zamanla nasıl değiştiğini karşılaştırmak
için line chart kullanılabilir. Her bir değişkeni farklı renkler veya çizgilerle temsil ederek, trendlerin ve
ilişkilerin gözlemlenmesini sağlar.

* Eğilim ve Desen Analizi: Line chart, bir değişkenin zaman içindeki trendini veya desenini belirlemek için
kullanılabilir. Örneğin, bir ürünün satışlarının mevsimsel bir deseni olup olmadığını belirlemek için kullanılabilir.

* Grafiksel Temsil: Veri setinin yapısal özelliklerini görsel olarak temsil etmek için kullanılır. Line chart,
veri setindeki değişkenlerin genel eğilimlerini ve ilişkilerini gösterir.

Özetlemek gerekirse, line chart, sürekli bir değişkenin zaman, sıralama veya başka bir sürekli değişken üzerindeki
değişimini gösteren önemli bir grafik türüdür.

```python
years = [2006 + x for x in range(16)]
weights = [random.randint(80, 90) + x for x in range(16)]
```

```python
plt.plot(years, weights, c="b", lw=4, linestyle="--")
plt.xlabel("Years")
plt.ylabel("Weights")
plt.title("Change in weight over time")
```

## Bar chart

Bar chart, kategorik verilerin (örneğin, kategoriler, gruplar veya sınıflar) frekansını veya dağılımını görsel
olarak temsil eden bir grafiktir. Bar chart'ın temel kullanımları şunlardır:

* Kategorilerin Karşılaştırılması: Farklı kategorilerin (örneğin, ürünler, bölgeler, zaman dilimleri vb.) sayısal
değerlerini karşılaştırmak için kullanılır. Örneğin, bir restoranın farklı yemek kategorilerinden ne kadar satış
yaptığını göstermek için bar chart kullanılabilir.

* Sıklık Dağılımı Gösterme: Belirli bir kategorideki sıklıkları veya frekansları göstermek için kullanılır. Örneğin,
bir anket sonuçlarındaki farklı yanıt kategorilerinin frekanslarını göstermek için kullanılabilir.

* Zaman Dilimlerindeki Değişiklikleri Gösterme: Belirli bir zaman dilimindeki kategorik verilerin değişimini göstermek
için kullanılabilir. Örneğin, aylık satış verilerini kategorik olarak temsil ederek her ayın satışlarını
karşılaştırmak için kullanılabilir.

* Toplamları Gösterme: Kategorik verilerin toplamını veya birleşik değerlerini göstermek için kullanılabilir. Örneğin,
farklı ürün kategorilerinin toplam satış değerlerini göstermek için kullanılabilir.

* Karar Vermeye Yardımcı Olma: Farklı seçeneklerin veya kategorilerin performansını karşılaştırarak karar verme
sürecini kolaylaştırır.

* Dağılımın Görsel Temsili: Belirli bir kategori içindeki değerlerin nasıl dağıldığını gösterir.

Özetlemek gerekirse, bar chart, kategorik verilerin frekansını veya dağılımını görsel olarak temsil eden önemli bir
grafik türüdür.

```python
x = ["Python", "Javascript", "Java", "C++", "Go"]
y = [random.randint(0, 200) * 1 for _ in range(5)]
```

```python
plt.bar(x, y, color="green", width=0.5)
plt.xlabel("Languages")
plt.ylabel("Number of users")
```

## Histogram

Histogram, sürekli bir değişkenin veri dağılımını görsel olarak temsil eden bir grafiktir. Histogramın temel
kullanımları şunlardır:

* Veri Dağılımını Gösterme: Histogram, bir sürekli değişkenin değerlerinin hangi aralıklarda ne sıklıkta olduğunu
gösterir. Bu, veri setinin genel dağılımı hakkında bilgi sağlar.

* Merkezi Eğilimi ve Dağılımı Gösterme: Histogram, veri setinin merkezi eğilimini (örneğin, ortalama veya medyan) ve
dağılımını (örneğin, varyans veya standart sapma) göstermek için kullanılır.

* Aykırı Değerlerin Belirlenmesi: Histogram, veri setindeki aykırı değerleri belirlemek için kullanılabilir. Aykırı
değerler, genellikle diğer değerlerden belirgin bir şekilde farklı olan ve analizi etkileyebilecek değerlerdir.

* Çoklu Veri Setlerini Karşılaştırma: Farklı veri setlerinin dağılımlarını karşılaştırmak için kullanılabilir. Bu,
farklı gruplar arasındaki benzerlikleri veya farkları belirlemek için önemlidir.

* Olabilirlik ve Frekans Dağılımı Analizi: Olabilirlik (probability) ve frekans dağılımı analizi yapmak için
kullanılabilir. Özellikle, belirli bir değerin olasılığını veya frekansını belirlemek için kullanılabilir.

* Normal Dağılım Kontrolü: Bir değişkenin normal dağılıma uygun olup olmadığını değerlendirmek için kullanılabilir.
Normal dağılım, birçok istatistiksel analiz için önemlidir.

Histogram, veri setinin genel yapısını ve karakteristiğini anlamak için önemli bir araçtır.

```python
ages = np.random.normal(20, 1.5, 1000)
#plt.hist(ages, bins=[ages.min(), 16, 18, 20, ages.max()])
plt.hist(ages, bins=20, cumulative=True)
```

## Pie Chart

Pasta grafiği , bir bütünün parçalarının oransal olarak gösterildiği bir veri
görselleştirmesi türüdür. Pasta grafiği, toplamın her bir parçasının yüzdesini veya oranını göstermek için
kullanılır. Bu grafik türü, veri setinin bütününü temsil etmek ve farklı kategorilerin oranlarını görsel olarak
karşılaştırmak için kullanılır.

Pasta grafiğinin temel kullanım alanları şunlardır:

* Oransal Karşılaştırma: Pasta grafiği, farklı kategorilerin bütün içindeki oransal payını gösterir. Her dilim,
bir kategoriyi temsil eder ve dilimin alanı veya açısı, bu kategorinin oransal büyüklüğünü belirtir.

* Toplamın İzlenmesi: Pasta grafiği, toplamın ne kadar olduğunu kolayca gösterir. Bütün dilimlerin toplamı, genellikle
100% veya 360 derece olur.

* Görsel Anlatım: Pasta grafiği, veri setinin yapısını hızlı bir şekilde anlamak için etkili bir görsel araçtır. Farklı
kategorilerin büyüklüklerini göz önüne alarak, gözlemlerin oransal dağılımını hemen anlayabilirsiniz.

* Basitlik ve Anlaşılabilirlik: Pasta grafiği, basit ve anlaşılır bir şekilde veri setinin ana bileşenlerini gösterir.
Bu nedenle, hızlı bir bakışta farklı kategorilerin önemini kavramak kolaydır.

* Sınırlamalar: Ancak, çok sayıda kategori veya çok küçük oranlara sahip kategorilerin olduğu durumlarda, pasta grafiği
kullanmak tavsiye edilmez. Bu tür durumlarda, bilgiyi etkili bir şekilde iletmek için diğer grafik türleri (örneğin,
bar chart veya line chart) daha uygundur.

Özetlemek gerekirse, pasta grafiği, farklı kategorilerin toplam içindeki oransal payını görsel olarak temsil eden
etkili bir veri görselleştirme aracıdır.

```python
langs = ["Python", "Javascript", "Java", "C++", "Go"]
votes = [50, 24, 36, 16, 32]
explodes = [0.2, 0, 0, 0, 0]
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%")
```

## Box plot

Box plot (kutu grafiği veya kutu çizimi), veri setinin istatistiksel dağılımını ve merkezi eğilimini görsel olarak
temsil eden bir grafik türüdür. Box plot, beş temel istatistiksel ölçümü gösterir: minimum, birinci çeyrek (Q1),
medyan, üçüncü çeyrek (Q3) ve maksimum değer.

Box plot'ın temel kullanım alanları şunlardır:

* Merkezi Eğilimi Gösterme: Box plot, veri setinin merkezi eğilimini belirtir. Kutunun ortasındaki yatay çizgi,
veri setinin medyanını gösterir.

* Çeyreklerin ve İnterquartile Range (IQR) Gösterimi: Box plot, veri setinin çeyreklerini (Q1, Q3) gösterir. Kutunun
üst ve alt kenarları, Q1 ve Q3'ü temsil eder. Bu, veri setinin ortalamasının dağılımını anlamanıza yardımcı olur.

* Aykırı Değer Tespiti: Box plot, potansiyel aykırı değerleri belirlemek için kullanılabilir. Aykırı değerler,
genellikle kutunun dışında kalan noktalardır.

* Dağılımın Gösterimi: Box plot, veri setinin dağılımını gösterir. Kutunun boyutu, veri setinin genel yayılımını temsil
eder.

* Veri Gruplarının Karşılaştırılması: Birden fazla veri grubunu karşılaştırmak için kullanılabilir. Bu,
farklı kategoriler veya gruplar arasındaki merkezi eğilim ve dağılım farklarını anlamak için önemlidir.

* Normal Dağılımın Kontrolü: Box plot, bir değişkenin normal dağılıma uygun olup olmadığını değerlendirmek için
kullanılabilir.

* Outlier Tespiti: Aykırı değerlerin belirlenmesi için kullanışlıdır. Aykırı değerler, genellikle kutunun dışında kalan
noktalardır.

Özetlemek gerekirse, box plot, veri setinin istatistiksel özelliklerini görsel olarak temsil eden önemli bir grafik
türüdür ve veri analizi ve karşılaştırmalar için kullanılır.

```python
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
```
